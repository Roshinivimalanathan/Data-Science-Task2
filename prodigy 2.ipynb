```python
data.dropna(subset = ['Age','Cabin','Embarked'], inplace=True)
data
```

```python
plt.figure(figure=(8,6))
explode=[0,0.1]
colors=['lavender','pink']
plt.pie(data['Survived'].value_counts(), labels=data['Survived'].value_counts().index, autopct='%1.1f%%',explode=explode,colors=colors)
plt.title('Survival rate')
plt.show() 
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data, x='Age', bins=20, kde=True, color='violet')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.boxplot(data, x='SibSp', y='Fare', palette='pink')
plt.xlabel('SibSp')
plt.ylabel('Fare')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
# Box plot comparing fares by passenger class
plt.figure(figsize=(5, 3))
sns.boxplot(data, x='Pclass', y='Age', palette='Blues')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
# Pair plot for numerical variables
sns.pairplot(data[['Age', 'Fare', 'Pclass', 'Survived']], hue='Survived', palette='pastel')
plt.title('Pair Plot of Numerical Variables')
plt.show()
```
